```python
# Filter rows with NaN values in the 'death_rate' column
nan_death_rate_countries = df[df['infant_mortality_rate'].isna()]

# Display the countries with NaN values for death_rate
print("Countries with NaN values for infant_mortality_rate:")
print(nan_death_rate_countries[['country', 'infant_mortality_rate']])

# Explore other aspects of these countries
print("\nMain aspects of these countries:")
print(nan_death_rate_countries.describe(include='all'))
```

**TASK: Notice climate is missing for a few countries, but not the Region! Let's use this to our advantage. Fill in the missing Climate values based on the mean climate value for its region.**
<br>
*NOTE! using death rate in this context, region is unavailable in this dataset, so I am using country*
<br>
Hints on how to do this: https://stackoverflow.com/questions/19966018/pandas-filling-missing-values-by-mean-in-each-group


```python
# Fill missing death_rate values based on mean death_rate for each country
df['death_rate'] = df.groupby('country')['death_rate'].transform(lambda x: x.fillna(x.mean()))

# Fill remaining NaN values with overall mean birth_rate
overall_mean_birth_rate = df['death_rate'].mean()
df['death_rate'].fillna(overall_mean_birth_rate, inplace=True)
df
```

**TASK:Use the same method for the missing values in Literacy.**
<br>
*NOTE! using life_exp_at_birth in this context*

```python
df['life_exp_at_birth'] = df.groupby('country')['life_exp_at_birth'].transform(lambda x: x.fillna(x.mean()))

overall_mean_birth_rate = df['life_exp_at_birth'].mean()
df['life_exp_at_birth'].fillna(overall_mean_birth_rate, inplace=True)
df
```

# 3. Data Feature Preparation

**TASK: Create a dummy variable to all categorical data.**

```python
# Convert 'country' column to dummy variables
dummy_variables = pd.get_dummies(df['country'], drop_first=True)

# Concatenate the dummy variables with the original DataFrame
df = pd.concat([df, dummy_variables], axis=1)

# Drop the original 'country' column
df.drop('country', axis=1, inplace=True)

df
```

**TASK: For numerical data perform standard scaler.**

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns
num_cols = df.select_dtypes(include=['float64']).columns

# Initialize StandardScaler
scaler = StandardScaler()

# Fit and transform the numerical columns
df[num_cols] = scaler.fit_transform(df[num_cols])

df
```

# 4. Creating and Fitting Kmeans Model

**TASK:** 

- Plot the within sum of squares vs. the number of cluster and find the optimal number of clusters.
- Fit a k-means model of the optimum cluster.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer

# Select the numerical columns for clustering
X = df.select_dtypes(include=['float64'])

# Initialize SimpleImputer with strategy='mean' to impute missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Impute missing values in the numerical columns
X_imputed = imputer.fit_transform(X)

# Convert the imputed array back to a DataFrame
X_imputed_df = pd.DataFrame(X_imputed, columns=X.columns)


# Initialize an empty list to store the within-cluster sum of squares
wcss = []

# Try different numbers of clusters and compute WCSS for each
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_imputed_df)
    wcss.append(kmeans.inertia_)

# Plot the WCSS vs. number of clusters
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Initialize PCA with 2 components for visualization
pca = PCA(n_components=2)

# Fit PCA to the data and transform it
X_pca = pca.fit_transform(X_imputed_df)

# Fit a KMeans model with the optimal number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans.fit(X_imputed_df)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Visualize the clusters in a scatter plot with a pink colormap
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='pink', edgecolor='k')
plt.title('KMeans Clustering with Optimal Number of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

# 5. Model Interpretation


**TASK: What K value do you think is a good choice? Are there multiple reasonable choices? What features are helping define these cluster choices.**

<p>Determining the optimal number of clusters (K) and understanding the characteristics of the resulting clusters are crucial stages in interpreting the KMeans clustering model. In this analysis, the elbow method was employed to determine the appropriate K value. By plotting the within-cluster sum of squares (WCSS) against various cluster numbers, we identified the point where the rate of decrease in WCSS slows down, signifying diminishing returns in clustering enhancement. For the dataset under consideration, the elbow point was observed at K=3, suggesting that three clusters may adequately capture the underlying structure of the data. However, it's important to note that depending on the context of the analysis, there may be multiple valid options for K, making the selection process not always straightforward. Therefore, exploring other plausible K values, such as K=4 or K=5, is recommended to assess whether they provide additional meaningful insights into the data.</p>
<br>
<p>To understand the features that characterise these cluster choices, we can look at the cluster centroids. The centroid of a cluster represents the average value of each attribute within the cluster. We are able to identify the features that most influence the separation of clusters by analysing centroid values for a range of K values. Features with significantly different centroid values between clusters are likely to have an impact on cluster borders. Box plots and scatter plots can be used to visualise the distribution of characteristics across clusters, providing additional insights on the discriminative strength of each feature. Overall, a combination of statistical analysis and visual investigation is necessary to understand the KMeans clustering results and the underlying structure of the data.</p>
