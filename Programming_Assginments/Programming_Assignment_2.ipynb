```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'target']
iris_data = pd.read_csv(url, names=column_names)

# Scatter plot: Sepal Length vs. Sepal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=iris_data, palette='Set1', legend='full')
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

# Scatter plot: Petal Length vs. Petal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', data=iris_data, palette='Set1', legend='full')
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

# Histogram: Sepal Length Distribution
plt.figure(figsize=(8, 6))
sns.histplot(iris_data, x='sepal length (cm)', hue='target', element='step', kde=True, palette='Set1', legend=True)
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.show()

# Histogram: Petal Length Distribution
plt.figure(figsize=(8, 6))
sns.histplot(iris_data, x='petal length (cm)', hue='target', element='step', kde=True, palette='Set1', legend=True)
plt.title('Petal Length Distribution')
plt.xlabel('Petal Length (cm)')
plt.show()

# Histogram: Petal Width Distribution
plt.figure(figsize=(8, 6))
sns.histplot(iris_data, x='petal width (cm)', hue='target', element='step', kde=True, palette='Set1', legend=True)
plt.title('Petal Width Distribution')
plt.xlabel('Petal Width (cm)')
plt.show()
```
